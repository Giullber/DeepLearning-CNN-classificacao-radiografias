```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="whitegrid", rc=custom_params, 
              palette=["Green","Orange"], font_scale=1.5)
# Funo de ativao Heaviside
def heaviside(x):
    return np.piecewise(x, [x < 0, x >= 0], [0, 1])

# Funo de ativao Linear
def linear(x):
    return x

# Funo de ativao Linear por partes
def piecewise_linear(x):
    return np.piecewise(x, [x < -1, (x >= -1) & (x < 1), x >= 1], [-1, lambda x: x, -1])

# Funo de ativao Sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Funo de ativao Softmax (para mltiplas classes)
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / exp_x.sum(axis=0, keepdims=True)

# Funo de ativao ReLU (Rectified Linear Unit)
def relu(x):
    return np.maximum(0, x)

# Criando um intervalo de valores para x
x = np.linspace(-5, 5, 1000)

# Plotando os grficos
plt.figure(figsize=(15, 10))

# Funo Heaviside
plt.subplot(2, 3, 1)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, heaviside(x))
plt.title('Funo Heaviside')


# Funo Linear
plt.subplot(2, 3, 2)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, linear(x))
plt.title('Funo Linear')

# Funo Linear por partes
plt.subplot(2, 3, 3)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, piecewise_linear(x))
plt.title('Funo Linear por Partes')

# Funo Sigmoide
plt.subplot(2, 3, 4)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, sigmoid(x))
plt.title('Funo Sigmoide')

# Funo Softmax (para 3 classes)
plt.subplot(2, 3, 5)
x_softmax = np.array([1, 2, 3])
plt.plot(x_softmax, softmax(x_softmax))
plt.title('Funo Softmax')

# Funo ReLU
plt.subplot(2, 3, 6)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, relu(x))
plt.title('Funo ReLU')

# Ajustando o layout
plt.tight_layout()

# Exibindo os grficos
plt.show()
```

```python
import locale
from matplotlib.ticker import FuncFormatter
# Configurar a localidade para o Brasil (pt_BR)
locale.setlocale(locale.LC_ALL, 'pt_BR')

#### Funao de formatao para separa n;meros decimais por virgualas
def format_decimais(value, _):
    return locale.format_string("%.2f", value, grouping=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="whitegrid", rc=custom_params, 
              palette=["Green","Orange"], font_scale=1.5)
# Funo de ativao Heaviside
def heaviside(x):
    return np.piecewise(x, [x < 0, x >= 0], [0, 1])

# Funo de ativao Linear
def linear(x):
    return x

# Funo de ativao Linear por partes
def piecewise_linear(x):
    return np.piecewise(x, [x < -1, (x >= -1) & (x < 1), x >= 1], [-1, lambda x: x, -1])

# Funo de ativao Sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Funo de ativao Softmax (para mltiplas classes)
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / exp_x.sum(axis=0, keepdims=True)

# Funo de ativao ReLU (Rectified Linear Unit)
def relu(x):
    return np.maximum(0, x)

# Criando um intervalo de valores para x
x = np.linspace(-5, 5, 1000)

# Plotando os grficos
plt.figure(figsize=(15, 10))

# Funo Heaviside
plt.subplot(2, 3, 1)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, heaviside(x))
plt.title('a)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_decimais))
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_decimais))

# Funo Linear
plt.subplot(2, 3, 2)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, linear(x))
plt.title('b)')
# plt.gca().yaxis.set_major_formatter(FuncFormatter(format_decimais))
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_decimais))

# Funo Linear por partes
plt.subplot(2, 3, 3)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, piecewise_linear(x))
plt.title('c)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_decimais))
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_decimais))

# Funo Sigmoide
plt.subplot(2, 3, 4)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, sigmoid(x))
plt.title('d)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_decimais))
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_decimais))

# Funo ReLU
plt.subplot(2, 3, 5)
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='x=0')
plt.plot(x, relu(x))
plt.title('e)')
# plt.gca().yaxis.set_major_formatter(FuncFormatter(format_decimais))
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_decimais))

# Ajustando o layout
plt.tight_layout()

# Exibindo os grficos
plt.show()
```
